```python
board=[
    [5,3,0,0,7,0,0,0,0],
    [6,0,0,1,9,5,0,0,0],
    [0,9,8,0,0,0,0,6,0],
    [8,0,0,0,6,0,0,0,3],
    [4,0,0,8,0,3,0,0,1],
    [7,0,0,0,2,0,0,0,6],
    [0,6,0,0,0,0,2,8,0],
    [0,0,0,4,1,9,0,0,5],
    [0,0,0,0,8,0,0,7,9]
]

def valid(b, num, pos):
    
    #check row
    for i in range(len(b)):
        if b[pos[0]][i]==num and i!=pos[1]:
            return False
    
    #check column
    for i in range(len(b)):
        if b[i][pos[1]]==num and i!=pos[0]:
            return False
    
    #check box
    x_box=pos[0] // 3
    y_box=pos[1] // 3
    
    for i in range( 3*x_box, 3*x_box+3):
        for j in range( 3*y_box, 3*y_box+3):
            if b[i][j]==num and (i,j)!=pos:
                return False
    return(True)
        
        
def print_board(b):
    for i in range(len(b)):
        if i%3==0 and i!=0:
            print("- - - - - - - - - -")
             
        for j in range(len(b)):
            if j%3==0 and j!=0:
                print("|", end="")
            print(b[i][j], end=" ")
        print("")
        
def find_empty(b):
    for i in range(len(b)):
        for j in range(len(b)):
            if b[i][j]==0:
                return(i,j)
    return None

def solve(b):
    find=find_empty(board)
    if  not find:
        return True
    else:
        for i in range(1,10):
            if valid(board,i,find):
                board[find[0]][find[1]]=i
                
                if solve(board):
                    return True
                
                board[find[0]][find[1]]=0
    return False
           
print_board(board)    
solve(board)
print("Solved Sudoku")
print_board(board)
```
